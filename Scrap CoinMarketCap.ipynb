```python
# Necessary to import
import urllib as ul
import pandas as pd
import numpy as np
from bs4 import BeautifulSoup

# Test import
import datetime
import threading
import time
```

```python
Bitcoin_URL = "https://coinmarketcap.com/currencies/bitcoin/historical-data/?start=20130428&end=20170926"
Ethereum_URL = "https://coinmarketcap.com/currencies/ethereum/historical-data/?start=20130428&end=20170926"
Ripple_URL = "https://coinmarketcap.com/currencies/ripple/historical-data/?start=20130428&end=20170926"
Dash_URL = "https://coinmarketcap.com/currencies/dash/historical-data/?start=20130428&end=20170926"
Litecoin_URL = "https://coinmarketcap.com/currencies/litecoin/historical-data/?start=20130428&end=20170926"
```

```python
Bitcoin = ul.request.urlopen(Bitcoin_URL)
soup = BeautifulSoup(Bitcoin, 'html.parser')

columns = ['Date','Open','High','Low','Close','Volume','Market Cap']

divs = soup.findAll('td')

dates  = [divs[i].text for i in range(0,len(divs),7)]
open = [divs[i].text for i in range(1,len(divs),7)]
high = [divs[i].text for i in range(2,len(divs),7)]
low = [divs[i].text for i in range(3,len(divs),7)]
close = [divs[i].text for i in range(4,len(divs),7)]
volume = [divs[i].text for i in range(5,len(divs),7)]
MarketCap = [divs[i].text for i in range(6,len(divs),7)]

dict = {'dates' : dates, 'open' : open, 'high' : high }
df_b = pd.DataFrame([dates,open,high,low,close,volume,MarketCap],index = columns).T

df_b['Crypto'] = 'Bitcoin'
```
