```python
# Necessary to import
import urllib as ul
import pandas as pd
import numpy as np
from bs4 import BeautifulSoup

# Test import
import datetime
import threading
import time
```

Download daily information about some cryptocurrencies from coinmarketcap.com:

```python
Bitcoin_URL = "https://coinmarketcap.com/currencies/bitcoin/historical-data/?start=20130428&end=20170926"
Ethereum_URL = "https://coinmarketcap.com/currencies/ethereum/historical-data/?start=20130428&end=20170926"
Ripple_URL = "https://coinmarketcap.com/currencies/ripple/historical-data/?start=20130428&end=20170926"
Dash_URL = "https://coinmarketcap.com/currencies/dash/historical-data/?start=20130428&end=20170926"
Litecoin_URL = "https://coinmarketcap.com/currencies/litecoin/historical-data/?start=20130428&end=20170926"
coinmarketcap_URLs = {"Bitcoin": Bitcoin_URL, "Ethereum": Ethereum_URL, "Ripple": Ripple_URL, "Dash": Dash_URL,
                      "Litecoin": Litecoin_URL}
```

```python
coinmarketcap_fields = ['Date','Open','High','Low','Close','Volume','Market Cap']
df_currency_list = []
for currency, url in coinmarketcap_URLs.items():
    url_request = ul.request.urlopen(url)
    soup = BeautifulSoup(url_request, 'html.parser')

    divs = soup.findAll('td')
    dates  = [divs[i].text for i in range(0,len(divs),7)]
    open_price = [divs[i].text for i in range(1,len(divs),7)]
    high = [divs[i].text for i in range(2,len(divs),7)]
    low = [divs[i].text for i in range(3,len(divs),7)]
    close_price = [divs[i].text for i in range(4,len(divs),7)]
    volume = [divs[i].text for i in range(5,len(divs),7)]
    MarketCap = [divs[i].text for i in range(6,len(divs),7)]

    ind_df = pd.DataFrame([dates,open_price,high,low,close_price,volume,MarketCap],index = coinmarketcap_fields).T
    ind_df["currency"] = currency
    df_currency_list.append(ind_df)
    
df_currency = pd.concat(df_currency_list)
df_currency.reset_index(drop=True, inplace=True)
df_currency["source"] = "coinmarketcap"
df_currency
```
